### Part 2: 15 pts Description of Time Series Classification models

#### A. 5 pts Select one classification model type. Describe how the model works. Why would each be a good or bad fit for this type of data?


k Nearest Neighbors:

The k-nearest neighbors (KNN) algorithm can be used for time series by replacing the Euclidean distance metric with the dynamic time warping (DTW) metric. The DTW measures similarity between two series  that may not align exactly in time, speed, or length.

KNN takes a unlabled observation and compares it to a population of labled observations. By finding the K nearest distances to the unlabeled observance, we can infer its class label. In time series, we can compute the distance between two time series sequences using dynamic time warping. The DTW finds an optimal alignment between two time series sequences through warping the time dimension or using other distance measures such as frechet distance.

DTW is good for classifying sequences that have different time lengths. It is simple, robust, and does not require extensive hyperparameter tuning. However, it can be computationally expensive because during classification, it compares each observance with all the other observances in the training set. Since our data is 1460 time steps long, it might be pretty expensive. 

#### B. 5 pts Select a second classification model type. Describe how the model works. Why would each be a good or bad fit for this type of data?


Timeseries Forest Classifier: 

It is kind of an extension of the random forest. It splits the series into random intervals, with random start positions and random lengths. Then certain features are extracted from each interval into a single vector. The features are summary statistics. Lastly, a decision tree is trained on the extracted features. This process is repeated until all trees have been built. New series are classified according to the class that is predicted by the most trees.
For the sktime TSF, we find mean, std and slope for each interval, concatenate to form new data set, and build decision tree on new data set.
Since we are looking at the statistics of each time series, I think it will be useful for this data because it will be computationally efficient and capture the trend and patterns in each series so that classification can be more accurate.

#### C. 5 pts Select third classification model type. Describe how the model works. Why would each be a good or bad fit for this type of data?

Random Interval Spectral Ensemble (RISE):

RISE is a frequency based time series classifier that uses a single time series interval per tree and it is trained using spectral features extracted from the series, instead of summary statistics as we previously saw. It selects a random interval of a series, where the length is a power of 2. It uses the whole series for the first tree. For the same interval on each series, it applies the series-to-series feature extraction transformers (Fitted auto-regressive coefficients, Estimated autocorrelation coefficients, and power spectrum coefficients). Then, it forms a new training set by combining the extracted features. Lastly, it trains a decision tree classifier.
The class probabilities are calculated as a proportion of base classifier votes. RISE is useful for long series such as audio becuase it controls the runtime by creating an adaptive model of the time  in order to build one tree. This may not be useful for this data because we do not have very large intervals. Also, since some of the class, the frequency is very similar, it may not be useful to use this

### Part 3: 50 pts Select one method. Model and examine results

#### A. 10 pts Select only one method. Fit your model. Feel free to adjust parameters or try a grid search (optional)

```python
from sktime.classification.interval_based import TimeSeriesForest
```

```python
tsf=TimeSeriesForest()
tsf.fit(X_train,y_train)
y_pred=tsf.predict(X_test)
```

```python
fi = tsf.feature_importances_
fig, ax = plt.subplots(1, figsize=plt.figaspect(0.25))
fi.plot(ax=ax)
ax.set(xlabel="Time", ylabel="Feature importance");
```

#### B. 10 pts Return the accuracy score of the train set and test set (suggestion to use .score()). Print the confusion matrix and classification report of the test set.

```python
print(tsf.score(X_train,y_train))
```

```python
print(tsf.score(X_test,y_test))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

I would use recall because the cost of acting and building a forecast model to classify patterns of high electrical use is low but the oppurtunity of passing it up and the penalty and rate for the supply multiplying and destroying my company's performance is high. Recall is the sensitivity or the fraction of relevant instances that were retrieved. It is the ratio of correctly predicted positive values to actual positive values. A high recall indicates a model returns most of the relevant results. If the model to classify patterns of high electrical usage is able to correctly indicate those who use more appliances or electricity and achieves a high recall, it will help the company more. While precision looks at the ratio of correctly predicted positive values to total predicted positive values. For a high precision model, if the number of true positives is high, then the number of false positive should be low. Precision looks at false positives- so it classifies those who don't use high electricity as they do use high electricity. Recall however looks at false negatives so the model classifies those who use a lot of electricity as they don't. We don't want this for our company. We don't want to risk classifying those who use a lot of electricity as they don't. Therefore, to achieve a high recall, we want to ensure the number of false negatives is as low as possible.

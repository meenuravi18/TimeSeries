# Load Dataset

```python
from sktime.datasets import load_airline

y = load_airline()

y.plot(title='Airline Passengers Over Time')
```

# Split y into Train and Test Groups

```python
from sktime.forecasting.all import temporal_train_test_split
```

```python
y_train, y_test = temporal_train_test_split(y)
```

```python
y_train.index.min(), y_train.index.max()
```

```python
y_test.index.min(), y_test.index.max()
```

# ARMA of ARIMA

```python
from sktime.forecasting.arima import ARIMA
```

Combine both AR lags and MA

```python
arma_model = ARIMA(order=(1,0,1))
arma_model.fit(y_train)
fh = list(range(1, 1+len(y_test)))
y_pred = arma_model.predict(fh=fh )
```

Plot the results where blue is actual and the red is your prediction

```python
y_train.plot()
y_pred.plot(color='red')
y_test.plot(color = 'black')
```

# Evaluate yhat

```python
from sktime.performance_metrics.forecasting import smape_loss
```

```python
smape_loss(y_test, y_pred )
```

# Try Higher Orders

```python
arma2_model = ARIMA(order=(2,0,2))
arma2_model.fit(y_train)
fh = list(range(1,1+len(y_test)))
y_pred = arma2_model.predict(fh=fh)
smape = smape_loss(y_test, y_pred)
print(f'SMAPE: {smape}')
y_train.plot()
y_pred.plot(color='red')
y_test.plot(color = 'black')
```

# Difference the data

```python
differenced_data = (y-y.shift()).iloc[1:]
```

```python
differenced_data.plot()
```

```python
from sktime.transformations.series import boxcox
bctransformer = boxcox.BoxCoxTransformer()
diff_trans_data = bctransformer.fit_transform(differenced_data)
```

```python
diff_trans_data.plot()
```
